```python
# Mapping for binary or ordinal columns
df['Transmission'] = df['Transmission'].map({'Manual': 0, 'Automatic': 1})
df['Owner_Type'] = df['Owner_Type'].map({'First': 1, 'Second': 2, 'Third': 3, 'Fourth & Above': 4})

# Label encoding for nominal (non-ordinal) columns
location_mapping = {location: idx for idx, location in enumerate(df['Location'].unique())}
fuel_type_mapping = {fuel: idx for idx, fuel in enumerate(df['Fuel_Type'].unique())}

# Apply mappings for Location and Fuel_Type
df['Location'] = df['Location'].map(location_mapping)
df['Fuel_Type'] = df['Fuel_Type'].map(fuel_type_mapping)

# Display the transformed DataFrame
print(df)
```

```python
x = df.drop('Price', axis=1)
y = df['Price']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state = 3)
```

```python
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
model = LinearRegression(fit_intercept=True, positive=True)
```

```python
model.fit(x_train,y_train)
```

```python
predictions = model.predict(x_test)
```

```python
print(predictions[:5])
```

```python
from sklearn.metrics import accuracy_score, precision_score, classification_report                       
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test, predictions)
print(f"Mean Absolute Error: {mae:.2f}")
```

```python

from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, predictions)
print(f"Mean Squared Error: {mse:.2f}")
```

```python
rmse = mean_squared_error(y_test, predictions, squared=False)
print(f"Root Mean Squared Error: {rmse:.2f}")
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, predictions)
print(f"R-squared: {r2:.2f}")
```

```python
from sklearn.tree import DecisionTreeRegressor
dt_model = DecisionTreeRegressor()
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'max_depth':[None,10,20,30,40,50],
    'min_samples_split':[2,10,20],
    'min_samples_leaf':[1,5,10],
    'max_features':['sqrt','log2']
    
}
```

```python
grid_search = GridSearchCV(estimator=dt_model,param_grid=param_grid, cv=5, scoring='neg_mean_squared_error')  
grid_search.fit(x_train,y_train)
```

```python
best_dt_model = grid_search.best_estimator_
print("Best parameters:", grid_search.best_params_)
```

```python
from sklearn.metrics import mean_squared_error, r2_score 
y_pred = best_dt_model.predict(x_test)
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```
